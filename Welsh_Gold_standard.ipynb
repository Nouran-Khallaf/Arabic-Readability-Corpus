```python
def count_instances(dataframe):
  ## remove spaces at the beginig of words
  dataframe = dataframe.apply(lambda x : x.str.lstrip())
  ## remove spaces at the end of the words
  dataframe = dataframe.apply(lambda x : x.str.rstrip())
  dataframe = dataframe.apply(lambda x : x.str.strip())
  ## lowercae all instances 
  dataframe = dataframe.apply(lambda x : x.str.lower())
  words = dataframe[['Gair','Saesneg']].drop_duplicates().values.tolist()
  instance_count = pd.DataFrame()
  for word,sense in words:
    ##Select a subset oof the data based on the search word and its sense
    subset =dataframe[(dataframe['Gair']==word) & (dataframe['Saesneg']== sense.replace('\\xa0', ' '))]
    subset = subset[['Gair','Cyfystyr1','Cyfystyr2','Cyfystyr3','Cyfystyr4','Cyfystyr5','Cyfystyr6','Cyfystyr7','Cyfystyr8','Cyfystyr9','Cyfystyr10']]
    subset_transpose =subset.transpose()
    subset_transpose.columns = ["user_1", "user_2", "user_3",'user_4','user_5','user_6','user_7']
    ###drop the gair row
    subset_transpose.drop(subset_transpose.index[0], inplace=True)
    ##count how many time each instance appeared
    subset_transpose = subset_transpose.apply(pd.Series.value_counts)
    ## Get the final count
    instances = pd.DataFrame(subset_transpose.count(axis='columns'))
    instances.columns = ["count"]
    instances_count =  pd.DataFrame(instances['count'].value_counts().reset_index())
    instances_count.columns = ['Agreement', 'counts']
    instance_count = instance_count.append(instances_count)
  
  return (instance_count['Agreement'].value_counts().reset_index())
```

```python
### Creat a new subset merging the sense with its POS
dataframe_POS=pd.DataFrame()
## remove spaces at the beginig of words
data = data.apply(lambda x : x.str.lstrip())
## remove spaces at the end of the words
data = data.apply(lambda x : x.str.rstrip())
data = data.apply(lambda x : x.str.strip())
## lowercae all instances 
data = data.apply(lambda x : x.str.lower())
dataframe_POS['Gair']=data["Gair"]
dataframe_POS['Saesneg']=data["Saesneg"]
dataframe_POS["Cyfystyr1"] = data["Cyfystyr1"].str.cat(data["RhY1"], sep = "-")
dataframe_POS["Cyfystyr2"] = data["Cyfystyr2"].str.cat(data["RhY2"], sep = "-")
dataframe_POS["Cyfystyr3"] = data["Cyfystyr3"].str.cat(data["RhY3"], sep = "-")
dataframe_POS["Cyfystyr4"] = data["Cyfystyr4"].str.cat(data["RhY4"], sep = "-")
dataframe_POS["Cyfystyr5"] = data["Cyfystyr5"].str.cat(data["RhY5"], sep = "-")
dataframe_POS["Cyfystyr6"] = data["Cyfystyr6"].str.cat(data["RhY6"], sep = "-")
dataframe_POS["Cyfystyr7"] = data["Cyfystyr7"].str.cat(data["RhY7"], sep = "-")
dataframe_POS["Cyfystyr8"] = data["Cyfystyr8"].str.cat(data["RhY8"], sep = "-")
dataframe_POS["Cyfystyr9"] = data["Cyfystyr9"].str.cat(data["RhY9"], sep = "-")
dataframe_POS["Cyfystyr10"] = data["Cyfystyr10"].str.cat(data["RhY10"], sep = "-")
```

```python
def count_instances_POS(dataframe):
  words = dataframe[['Gair','Saesneg']].drop_duplicates().values.tolist()
  instance_count = pd.DataFrame()
  #subset_transpose_csv= pd.DataFrame()
  for word,sense in words:
    ##Select a subset oof the data based on the search word and its sense
    subset =dataframe[(dataframe['Gair']==word) & (dataframe['Saesneg']== sense.replace('\\xa0', ' '))]
    subset_transpose =subset.transpose()
    subset_transpose.columns = ["user_1", "user_2", "user_3",'user_4','user_5','user_6','user_7']
    ###drop the gair row
    subset_transpose.drop(subset_transpose.index[0], inplace=True)
    subset_transpose.drop(subset_transpose.index[0], inplace=True)
    ##count how many time each instance appeared
    subset_transpose = subset_transpose.apply(pd.Series.value_counts)
    ## Get the final count
    #subset_transpose_csv= subset_transpose_csv.append(subset_transpose)
    instances = pd.DataFrame(subset_transpose.count(axis='columns'))
    instances.columns = ["count"]
    instances_count =  pd.DataFrame(instances['count'].value_counts().reset_index())
    instances_count.columns = ['Agreement', 'counts']
    instance_count = instance_count.append(instances_count)
  #subset_transpose_csv.to_csv('subset_transpose_csv.csv') 
  return (instance_count['Agreement'].value_counts().reset_index())
```
